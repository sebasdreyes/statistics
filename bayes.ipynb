```python
# Install necessary libraries

import numpy as np
import pymc3 as pm
import matplotlib.pyplot as plt

# 1. Simulate some synthetic data
np.random.seed(42)
n_samples = 100
X = np.linspace(0, 1, n_samples)
true_slope = 2
true_intercept = 1
true_sigma = 0.1
y = true_intercept + true_slope * X + np.random.normal(0, true_sigma, n_samples)

# 2. Bayesian Linear Regression Model
with pm.Model() as model:
    # Priors for the intercept, slope, and sigma
    intercept = pm.Normal('intercept', mu=0, sigma=10)
    slope = pm.Normal('slope', mu=0, sigma=10)
    sigma = pm.HalfNormal('sigma', sigma=1)
    
    # Linear model
    y_est = intercept + slope * X
    
    # Likelihood function (observed data)
    likelihood = pm.Normal('y', mu=y_est, sigma=sigma, observed=y)
    
    # 3. Posterior inference
    trace = pm.sample(2000, return_inferencedata=False, tune=1000)

# 4. Plotting the results
pm.traceplot(trace)
plt.show()

# 5. Posterior estimates
intercept_post = np.mean(trace['intercept'])
slope_post = np.mean(trace['slope'])
sigma_post = np.mean(trace['sigma'])

print(f"Posterior Intercept: {intercept_post}")
print(f"Posterior Slope: {slope_post}")
print(f"Posterior Sigma: {sigma_post}")
```
